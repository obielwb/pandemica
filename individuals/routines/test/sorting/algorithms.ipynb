```python
import os

import pandas as pd
import matplotlib.pyplot as plt


algorithms = os.listdir('./data')

plt.figure(figsize=(10, 6))

iterations = -(2 ** 10)

for algorithm in algorithms:
    algorithm_data = pd.read_csv(f'./data/{algorithm}')
    
    plt.plot(algorithm_data['iteration'], algorithm_data['runtime'],
             label=f'{algorithm.split(".")[0].capitalize()} Sort', marker='o')
    
    if len(algorithm_data) > iterations:
        iterations = len(algorithm_data)

plt.xlabel('Iteração')
plt.xticks([x for x in range(iterations)])
plt.ylabel('Tempo de execução (s)')
plt.title('Algoritmos de ordenação')
plt.legend()
plt.show()
```
